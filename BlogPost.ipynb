# 3. Data Preparation 

```python
# Fill in missing data with zeros
ebola.fillna(0, inplace = True)
ebola_db.fillna(0, inplace = True)
```

```python
# See what countries are in the dataset
ebola['Country'].value_counts()
```

```python
# Count the necessary values for the first question
africa_2014 = 0
africa_2015 = 0
africa_2016 = 0
europe_2014 = 0
europe_2015 = 0
europe_2016 = 0
north_america_2014 = 0
north_america_2015 = 0
north_america_2016 = 0
for index, row in ebola.iterrows():
    date = ebola['Date'][index]
    country = ebola['Country'][index]
    date = re.split('[/-]', date)
    date = date[0]
    if (country == 'Liberia' or country == 'Guinea' or country == 'Sierra Leone' or country == 'Nigeria' or country == 'Senegal' or country == 'Mali') and date == '2014':
        africa_2014 = africa_2014 + ebola['No. of confirmed, probable and suspected deaths'][index]
    if (country == 'Italy' or country == 'Spain' or country == 'United Kingdom') and date == '2014':
        europe_2014 = europe_2014 + ebola['No. of confirmed, probable and suspected deaths'][index]
    if (country == 'United States of America') and date == '2014':
        north_america_2014 = north_america_2014 + ebola['No. of confirmed, probable and suspected deaths'][index]
    if (country == 'Liberia' or country == 'Guinea' or country == 'Sierra Leone' or country == 'Nigeria' or country == 'Senegal' or country == 'Mali') and date == '2015':
        africa_2015 = africa_2015 + ebola['No. of confirmed, probable and suspected deaths'][index]
    if (country == 'Italy' or country == 'Spain' or country == 'United Kingdom') and date == '2015':
        europe_2015 = europe_2015 + ebola['No. of confirmed, probable and suspected deaths'][index]
    if (country == 'United States of America') and date == '2015':
        north_america_2015 = north_america_2015 + ebola['No. of confirmed, probable and suspected deaths'][index]
    if (country == 'Liberia' or country == 'Guinea' or country == 'Sierra Leone' or country == 'Nigeria' or country == 'Senegal' or country == 'Mali') and date == '2016':
        africa_2016 = africa_2016 + ebola['No. of confirmed, probable and suspected deaths'][index]
    if (country == 'Italy' or country == 'Spain' or country == 'United Kingdom') and date == '2016':
        europe_2016 = europe_2016 + ebola['No. of confirmed, probable and suspected deaths'][index]
    if (country == 'United States of America') and date == '2016':
        north_america_2016 = north_america_2016 + ebola['No. of confirmed, probable and suspected deaths'][index]
```

```python
# Add all the years for the first question
year_2014 = africa_2014 + europe_2014 + north_america_2014
year_2015 = africa_2015 + europe_2015 + north_america_2015
year_2016 = africa_2016 + europe_2016 + north_america_2016
```

# 4. Analysis and Solutions

### Question 1. What year had the highest percentage for suspected, probable, and confirmed Ebola deaths?

```python
# This is to create a pie chart for the suspected, probable, and confirmed Ebola deaths for each year
plt.pie([year_2014, year_2015, year_2016], explode = [0, 0.1, 0], colors = ['blue', 'red', 'yellow'], autopct = '%1.1f%%', shadow = True, labels = ['2014', '2015', '2016'])
plt.axis('equal')
plt.title('Suspected, Probable, and Confirmed Ebola Deaths from 2014 - 2016')
plt.show()
```

#### The above pie chart shows that the majority of suspected, probable, and confirmed Ebola deaths happened in 2015. Ninety-two percent of suspected, probable, and confirmed Ebola deaths occured in 2015 compared to 2014 and 2016. Most suspected, probable, and confirmed Ebola deaths occured in Africa from 2014 to 2016.
